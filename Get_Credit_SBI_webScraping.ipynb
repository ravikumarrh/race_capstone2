```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# def cards_list(card_names):
#     for i in range(len(card_names)):
#         card = card_names[i].text
#         cards.append(card)
#     return cards

# def cards_list(card_names):
#     for i in range(len(card_names)):
#         for c in card_names[i].find('h4'):
#             card = c.text
#             cards.append(card)
#     return cards

def cards_list(card_names):
    for i in range(len(card_names)):
        for c in card_names[i].find_all('img',class_='lazy'):
            card = c.get('title')
            cards.append(card)
    return cards

def offers(offers_ul): 
    card_offers = [[] for _ in range(len(offers_ul))]
    for j in range(len(offers_ul)):   
        for offer in offers_ul[j].find_all("li",class_='',recursive=False):
            card_offers[j].append(offer.text.strip())
    return card_offers
```

```python
cards = []

# Making a GET request
r = requests.get('https://www.sbicard.com/en/personal/credit-cards.page')
# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')
#card_names = soup.find_all('h4')
card_names = soup.find_all('a')
card_list = cards_list(card_names)
cards
offers_ul = soup.find_all('ul',class_= '')
card_offers = offers(offers_ul)
card_offers
```

```python
# category = {'Category': card_category_list}
# #category = {'Category': card_category}
# df_card_category = pd.DataFrame.from_dict(category)

cards_data = {'Cards': cards}
df_cards = pd.DataFrame.from_dict(cards_data)
#df_cards

df_offers = pd.DataFrame.from_dict(list(card_offers))
#df_offers
df_offers.shape
list_of_offers = []
for i in range(1,len(df_offers. columns)+1):
    list_of_offers.append("Offer"+str(i))
list_of_offers
df_offers. columns = list_of_offers   

df_cards = pd.concat([df_cards, df_offers], axis=1)
df_cards
```
